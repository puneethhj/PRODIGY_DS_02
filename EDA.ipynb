```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Importing dataset 

```python
data = pd.read_csv("train.csv")
# type(data)
data.head(20)
data.columns
```

```python
data['Age'].fillna(data['Age'].median(), inplace=True)
data['Cabin'].fillna("Not Known", inplace=True)
data['Survived'].replace(1, "Yes", inplace=True)
data['Survived'].replace(0, "No", inplace=True)

data.drop_duplicates(inplace=True)
data.head(20)
```

Survival Distribution

```python
fig = plt.figure(figsize=(5,5))
ax = sns.countplot(x='Survived', data=data, palette='Paired')
plt.title("Survival Distribution")

for i in ax.containers:
    ax.bar_label(i,)

plt.xlabel("Surived")
plt.ylabel("No. of passengers")
plt.show()
```

Gender Distribution

```python
fig = plt.figure(figsize=(5,5))

gender_counts = data['Sex'].value_counts()
gender_counts

plt.pie(gender_counts, labels = gender_counts.index, autopct='%.0f%%')

plt.title("Passenger Gender distribution")
plt.show()
```

Passenger Class Distribution

```python
fig = plt.figure(figsize=(5,5))
ax = sns.countplot(x="Pclass", data=data, palette="Paired")
for i in ax.containers:
    ax.bar_label(i,)

plt.title("Passenger Class distribution")
plt.xlabel("Class")
plt.ylabel("No. of passengers")
plt.show()
```
