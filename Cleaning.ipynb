```python
import pandas as pd
```

Importing Dataset

```python
data = pd.read_csv("test.csv")
# print(type(data))
print(data.shape)
```

```python
data.head(20)
# print(data.columns)
```

Checking for null values

```python
print(data.isnull().sum())
```

```python
data.describe()
data.info()
```

# Filling null values of age with median value


```python
data['Age'].fillna(data['Age'].median(), inplace=True)
data.head(20)

```

# Checking for duplicate values


```python
y = data.duplicated().sum()
if y != 0 :
    data.drop_duplicates(inplace =True)
else:
    print ("no duplicates")
    
```

# Replacing null values in cabin column

```python

data['Cabin'].fillna("Not Known", inplace=True)
data.head(10)
print(data.isnull().sum())
```

# Filling null values in fare column

```python

data['Fare'].fillna(data['Fare'].median(), inplace=True)
data.isnull().sum()
```
